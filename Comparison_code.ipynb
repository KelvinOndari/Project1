```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# === 1. Load dataset ===
df = pd.read_csv("/content/M2. module_2_data (1).csv")

# For demonstration: predict US stock returns using predictors METALS, OIL, and DXY
y = df["US_STK"]
X = df[["METALS", "OIL", "DXY"]]
X = sm.add_constant(X)

# === 2. Fit Ordinary Least Squares (LS) ===
ols_model = sm.OLS(y, X).fit()

# === 3. Compute weights for WLS from LS residuals ===
# Variance is proportional to squared residual
residuals = ols_model.resid
weights = 1 / (residuals**2 + 1e-6)   # add small constant to avoid divide-by-zero

# === 4. Fit Weighted Least Squares (WLS) ===
wls_model = sm.WLS(y, X, weights=weights).fit()

# === 5. Compute test statistics â€” use AIC for comparison ===
ols_aic = ols_model.aic
wls_aic = wls_model.aic

print("=== Model Comparison (Lower AIC = Better) ===")
print(f"OLS AIC: {ols_aic:.4f}")
print(f"WLS AIC: {wls_aic:.4f}")

if wls_aic < ols_aic:
    print("\nWLS performs better than LS based on AIC.")
else:
    print("\nLS performs better than WLS based on AIC.")

# Optional: print model summaries
print("\nOLS Summary:")
print(ols_model.summary())

print("\nWLS Summary:")
print(wls_model.summary())
```
